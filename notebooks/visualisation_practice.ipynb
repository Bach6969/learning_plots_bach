```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
spotify_df.head()
```

```python
avg_streams = spotify_df.groupby("Genre")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_streams 
```

```python
plt.bar(avg_streams.index, avg_streams.values)
plt.show()
```

```python
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")

plt.show()
```

```python
plt.Figure(figsize=(10, 6))
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha="right")

plt.show()
```

```python
plt.Figure(figsize=(10, 6))
plt.bar(avg_streams.index, avg_streams.values)

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
avg_artist_streams = spotify_df.groupby("Artist")["Total Streams (Millions)"].mean().sort_values(ascending=False)
avg_artist_streams 
```

```python
plt.figure(figsize=(10, 6))

special_artist = "Ed Sheeran"
colors = ["#1f77b4" if artist == special_artist else "#8d8d8d" for artist in avg_artist_streams.index]

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color=colors, edgecolor="black", width=0.6)
ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.xlabel("Artist")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Artist")
plt.xticks(rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

```python
spotify_df.columns
spotify_df.info()
```

```python
plt.scatter(spotify_df["Streams Last 30 Days (Millions)"], spotify_df["Total Streams (Millions)"])
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"], 
    spotify_df["Total Streams (Millions)"],
    color="#219e40",
    alpha = 0.7,
    edgecolor="black",
    s = 25
    )

plt.xlabel("Streams this Month")
plt.ylabel("Total Streams")
plt.title("No Apparent Correlation Between Monthly Listens and Total Streams")

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

featured_genre = "Pop"

colors = ["#219e40" if genre == featured_genre else "#8d8d8d" for genre in spotify_df["Genre"]]

plt.scatter(
    spotify_df["Streams Last 30 Days (Millions)"], 
    spotify_df["Total Streams (Millions)"],
    color=colors,
    alpha = 0.7,
    s = 35
    )

plt.xlabel("Streams this Month")
plt.ylabel("Total Streams")

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.set_title("Do Pop Songs Achieve Higher Monthly Listens or Total Streams?", fontsize = 18, pad=35)

for spine in ax.spines.values():
    spine.set_linewidth(0.5)
    spine.set_alpha(0.7)

plt.tight_layout()
plt.show()
```

```python
yearly_avg = spotify_df.groupby("Release Year")["Total Streams (Millions)"].mean().round(2)
yearly_avg
```

```python
plt.figure(figsize=(10, 6))
plt.plot(yearly_avg.index, yearly_avg.values, marker="o", color="#ff7f0e", linewidth=2)

plt.xlabel("Release Year")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Release Year")

ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.axvline(x=2015, color="#2ca02c", linestyle="--", linewidth=1.5, label="2015 - Spotify Hits 100M Users")

plt.ylim(0, 3500)

plt.show()
```

```python
category = "Platform Type"
values = "Total Streams (Millions)"

platform_totals = spotify_df.groupby(category)[values].sum()
platform_totals
```

```python
plt.figure(figsize=(8, 8))
plt.pie(
    platform_totals,
    labels=None,
    startangle=110,
    colors=["#1f77b4", "#2ca02c"]
    )

plt.title("Total Streams by Platform Type")

plt.legend(title="Platform Type", loc="upper left")

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(
    spotify_df["Streams Last 30 Days (Millions)"], 
    bins=20, 
    color="#b41f82", 
    edgecolor="black")

plt.show()
```

```python
plt.boxplot(
    [spotify_df[spotify_df["Platform Type"]=="Free"]["Skip Rate (%)"],
    spotify_df[spotify_df["Platform Type"]=="Premium"]["Skip Rate (%)"]] )

plt.xticks([1, 2], ["Free", "Premium"])
plt.ylabel("Skip Rate (%)")
plt.xlabel("Platform Type")
plt.title("Skip Rates by Platform Type")

plt.show()
```

```python
numeric_spotify_df = spotify_df.select_dtypes(include=[np.number])
corr = numeric_spotify_df.corr()

im = plt.imshow(corr, cmap="coolwarm", vmin=-1, vmax=1)
plt.colorbar(im, fraction=0.046, pad=0.04, label="Correlation Coefficient")

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha="right")
plt.yticks(range(len(corr.columns)), corr.columns)

plt.show()
```
